## Read and Parse a FASTA File

```python
from Bio import SeqIO

# Load FASTA file (can be local or downloaded)
fasta_file = "example.fasta"  # Replace with your file

for record in SeqIO.parse(fasta_file, "fasta"):
    print("ID:", record.id)
    print("Description:", record.description)
    print("Sequence:", record.seq)
    print("Length:", len(record.seq))
```

## Count Amino Acids in a Sequence

```python
from collections import Counter

sequence = "MKTFFVFLGILALFNAQA"  # Example protein sequence

print("Amino Acid Counts:")
aa_counts = Counter(sequence)
for aa, count in aa_counts.items():
    print(f"{aa}: {count}")
```

```python
import matplotlib.pyplot as plt

# Frequency bar plot
plt.bar(aa_counts.keys(), aa_counts.values())
plt.xlabel("Amino Acid")
plt.ylabel("Frequency")
plt.title("Amino Acid Composition")
plt.show()
```

## Visualize Sequence Length Distribution from a FASTE file


```python
seq_lengths = []
for record in SeqIO.parse(fasta_file, "fasta"):
    seq_lengths.append(len(record.seq))

plt.hist(seq_lengths, bins=30)
plt.xlabel("Sequence length")
plt.ylabel("Count")
plt.title("Proteins Sequence Length Distribution")
plt.show()
```
